```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel('C:/Day 23 Dataset/Bank_Personal_Loan_Modelling.xlsx', sheet_name=1)
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.drop(['ID'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
randomForestClassifier = RandomForestClassifier(n_estimators=1000,max_features=12,oob_score=True)
```

```python
randomForestClassifier.fit(data.drop(['Personal Loan'],axis=1),data['Personal Loan'])
```

```python
z = zip(data.columns,randomForestClassifier.feature_importances_)
z_sorted = sorted(z,key= lambda x : x[1])
for feature,importance in z_sorted:
    print(feature,'-',importance)
```

# As we can see only 3 column is importance Family, Income, Education

1. Family - 0.1511912696394559
2. Income - 0.3390535276599535
3. Education - 0.3677233871280927


```python
decisionTreeClassifier = DecisionTreeClassifier(max_depth=4)
```

```python
decisionTreeClassifier.fit(data[['Family','Income','Education']],data['Personal Loan'])
```

```python
plt.rcParams["figure.figsize"] = (300,100)
```

```python
tree.plot_tree(decisionTreeClassifier,filled = True)
```
